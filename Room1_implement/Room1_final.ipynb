```python
import numpy as np
import pandas as pd

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# import metrics to compute accuracy (Evaluate)
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
path='/content/drive/MyDrive/Dataset/our dataset/modarg_1n_geofence_metrics_mf_bssid.csv'
df=pd.read_csv(path)
df.head(30)
```

```python
df.isnull().values.any()
```

```python
df=df.dropna()
```

```python
X= pd.DataFrame(df.drop(['LOCATION'],axis=1))
Y=pd.DataFrame(df.LOCATION.values)
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
# Fit the label encoder to the labels and transform the labels
y = label_encoder.fit_transform(Y)
# Print the encoded labels
print(y)
print(y.shape)
```

```python

new= pd.DataFrame(X.drop(['WIFI_0_LEVEL','WIFI_1_LEVEL','WIFI_2_LEVEL','WIFI_3_LEVEL','WIFI_4_LEVEL','MAGNETIC_FIELD_uT'],axis=1))
Xnew=new.T.reset_index().values.tolist()
#print(Xnew)
path='/content/drive/MyDrive/Dataset/our dataset/unique_BSSID.csv'
all_unique_BSSID=pd.read_csv(path)
all_unique_BSSID=pd.DataFrame(all_unique_BSSID)
all_unique_BSSID=all_unique_BSSID.T.reset_index().values.tolist()

print(all_unique_BSSID)

#labeling
def label_encode (Xnew,all_unique_BSSID):
    for i, row in enumerate(Xnew):
        for j, element in enumerate(row):
            for k, value in enumerate(all_unique_BSSID):
                if element == value:
                    Xnew[i][j] = k
    X_final=pd.DataFrame(Xnew).T
    return(X_final)
X_final=label_encode(Xnew,all_unique_BSSID[0])
print(X_final)
```

```python
new_path='/content/drive/MyDrive/Dataset/our dataset/m1_only_after_label.csv'
X_after_edit=pd.read_csv(new_path)
X_after_edit.head(30)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_after_edit, y, test_size = 0.2, random_state = 99)
RF =RandomForestClassifier(max_features='auto',n_estimators=100,max_depth=None)
RF.fit(X_train, y_train)
y_prediction=RF.predict(X_test)
score_of_model=accuracy_score(y_test, y_prediction)
print('Accuracy score of the ', RF, ' = ',  score_of_model)
```
