```python
X= pd.DataFrame(df.drop(['LOCATION'],axis=1))
Y=pd.DataFrame(df.LOCATION.values)
print(X)
```

Label encoding for output

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
# Fit the label encoder to the labels and transform the labels
y = label_encoder.fit_transform(Y)
# Print the encoded labels
print(y)
print(y.shape)
```

Label for BSSID

```python

new= pd.DataFrame(X.drop(['WIFI_0_LEVEL','WIFI_1_LEVEL','WIFI_2_LEVEL','WIFI_3_LEVEL','WIFI_4_LEVEL','MAGNETIC_FIELD_uT'],axis=1))
Xnew=new.T.reset_index().values.tolist()
#print(Xnew)

w0=df["WIFI_0_BSSID"].unique()
w1=df["WIFI_1_BSSID"].unique()
w2=df["WIFI_2_BSSID"].unique()
w3=df["WIFI_3_BSSID"].unique()
w4=df["WIFI_4_BSSID"].unique()
result=np.concatenate((w0,w1,w2,w3,w4))
unique_BSSID=np.unique(result)
print(unique_BSSID)
print(unique_BSSID.shape)

#labeling
def label_encode (Xnew,unique_BSSID):
    for i, row in enumerate(Xnew):
        for j, element in enumerate(row):
            for k, value in enumerate(unique_BSSID):
                if element == value:
                    Xnew[i][j] = k
    X_final=pd.DataFrame(Xnew).T
    return(X_final)

X_final=label_encode (Xnew,unique_BSSID)
print(X_final)
```
